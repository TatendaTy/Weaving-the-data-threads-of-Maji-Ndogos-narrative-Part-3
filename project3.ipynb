#### Linking records to employees
Look at the errors that might have caused the 6% of records that were incorrect. At some of the locations, employees assigned scores incorrectly, and those records ended up in the result set.

Reasons for incorrect scores:
- Employees are humans and can make mistakes, which is something to consider when using human auditors to check water quality.

- The alternative is that someone assigned the scores incorrectly on purpose, which could be a sign of corruption or negligence. This is something that Maji Ndogo should investigate further to ensure the integrity of their water quality assessments.

```python
%%sql
-- 1. JOIN the assigned_employeed_id from the visits table to the employees table to get the employee names for each visit.
select
    v.record_id as 'Visit Record ID',
    v.location_id as 'Visit Location ID',
    v.assigned_employee_id as 'Assigned Employee ID',
    auditor_report.true_water_source_score as 'Auditor Score',
    water_quality.subjective_quality_score as 'Surveyor Score'
from visits v
join employee e on v.assigned_employee_id = e.assigned_employee_id
join auditor_report on v.location_id = auditor_report.location_id
join water_quality on v.record_id = water_quality.record_id
where v.visit_count = 1 and water_quality.subjective_quality_score != auditor_report.true_water_source_score;
```

#### Gathering evidence of employee errors

```python
%%sql
-- calculate the average number of mistakes employees made
with Error_count as (
    select
        employee_name,
        count(*) as 'Number of mistakes'
    from (
        select
            v.record_id as 'Visit Record ID',
            v.location_id as 'Visit Location ID',
            e.employee_name as employee_name,
            auditor_report.true_water_source_score as 'Auditor Score',
            water_quality.subjective_quality_score as 'Surveyor Score'
        from visits v
        join employee e on v.assigned_employee_id = e.assigned_employee_id
        join auditor_report on v.location_id = auditor_report.location_id
        join water_quality on v.record_id = water_quality.record_id
        where v.visit_count = 1 and water_quality.subjective_quality_score != auditor_report.true_water_source_score
    ) Incorrect_records
    group by employee_name
)
select avg(`Number of mistakes`) as avg_error_count_per_empl from Error_count;
```

```python
%%sql
-- compare each employee's error_count with avg_error_count_per_empl.
with Error_count as (
    select
        employee_name,
        count(*) as `Number of mistakes`
    from (
        select
            v.record_id as 'Visit Record ID',
            v.location_id as 'Visit Location ID',
            e.employee_name as employee_name,
            auditor_report.true_water_source_score as 'Auditor Score',
            water_quality.subjective_quality_score as 'Surveyor Score'
        from visits v
        join employee e on v.assigned_employee_id = e.assigned_employee_id
        join auditor_report on v.location_id = auditor_report.location_id
        join water_quality on v.record_id = water_quality.record_id
        where v.visit_count = 1 and water_quality.subjective_quality_score != auditor_report.true_water_source_score
    ) Incorrect_records
    group by employee_name
)
select
    employee_name,
    `Number of mistakes`,
    `Number of mistakes` > (select avg(`Number of mistakes`) from Error_count) as suspect
from Error_count
order by `Number of mistakes` desc;
```

```python
%%sql
-- create a view for our incorrect records
create or replace view Incorrect_records as
select
    ar.location_id as `Location ID`,
    v.record_id as `Visit Record ID`,
    e.employee_name as employee_name,
    ar.true_water_source_score as `Auditor Score`,
    wq.subjective_quality_score as `Surveyor Score`,
    ar.statements as `Statements`
from auditor_report ar
join visits v on ar.location_id = v.location_id
join water_quality wq on v.record_id = wq.record_id
join employee e on v.assigned_employee_id = e.assigned_employee_id
where v.visit_count = 1
  and ar.true_water_source_score != wq.subjective_quality_score;
```

```python
%%sql
-- calculate the average number of the number_of_mistakes in error_count
with Error_count as (
    select
        employee_name,
        count(*) as `Number of mistakes`
    from Incorrect_records
    group by employee_name
)
select avg(`Number of mistakes`) as avg_error_count_per_empl from Error_count;
```

```python
%%sql
-- compare each employee's error_count with avg_error_count_per_empl.
with Error_count as (
    select
        employee_name,
        count(*) as `Number of mistakes`
    from Incorrect_records
    group by employee_name
)
select
    employee_name,
    `Number of mistakes`
from Error_count
where `Number of mistakes` > (select avg(`Number of mistakes`) from Error_count);
```

```python
%%sql
-- get the records these employees gathered and inspect statement patterns
with Error_count as ( -- this CTE calculates the number of mistakes each employee made
    select
        employee_name,
        count(*) as `Number of mistakes`
    from Incorrect_records
        /* Incorrect_records is a view that joins the audit report to the database
        for records where the auditor and employees scores are different*/
    group by employee_name
),
Avg_error_count as ( -- this CTE calculates the average number of mistakes per employee
    select avg(`Number of mistakes`) as avg_error_count_per_empl
    from Error_count
),
Suspect_employees as (
    select employee_name
    from Error_count
    where `Number of mistakes` > (select avg_error_count_per_empl from Avg_error_count)
)
-- this query filter all the incorrect records to only include those recorded by the suspect employees, and inspect the statements they made when recording the water source quality.
select
    ir.employee_name,
    ir.`Location ID` as location_id,
    ir.`Statements` as statements
from Incorrect_records ir
join Suspect_employees se on ir.employee_name = se.employee_name;
```

```python
%%sql
-- filtering the statements to only include those that contain the word "cash"
with Error_count as ( -- this CTE calculates the number of mistakes each employee made
    select
        employee_name,
        count(*) as `Number of mistakes`
    from Incorrect_records
        /* Incorrect_records is a view that joins the audit report to the database
        for records where the auditor and employees scores are different*/
    group by employee_name
),
Avg_error_count as ( -- this CTE calculates the average number of mistakes per employee
    select avg(`Number of mistakes`) as avg_error_count_per_empl
    from Error_count
),
Suspect_employees as (
    select employee_name
    from Error_count
    where `Number of mistakes` > (select avg_error_count_per_empl from Avg_error_count)
)
-- this query filter all the incorrect records to only include those recorded by the suspect employees, and inspect the statements they made when recording the water source quality.
select
    ir.employee_name,
    ir.`Location ID` as location_id,
    ir.`Statements` as statements
from Incorrect_records ir
join Suspect_employees se on ir.employee_name = se.employee_name
where ir.`Statements` like '%cash%';
```

```python
%%sql
-- checking if there are any employees in the Incorrect_records table with statements mentioning "cash" that are not in our suspect list
with Error_count as ( -- this CTE calculates the number of mistakes each employee made
    select
        employee_name,
        count(*) as `Number of mistakes`
    from Incorrect_records
        /* Incorrect_records is a view that joins the audit report to the database
        for records where the auditor and employees scores are different*/
    group by employee_name
),
Avg_error_count as ( -- this CTE calculates the average number of mistakes per employee
    select avg(`Number of mistakes`) as avg_error_count_per_empl
    from Error_count
),
Suspect_employees as (
    select employee_name
    from Error_count
    where `Number of mistakes` > (select avg_error_count_per_empl from Avg_error_count)
),
Cash_statements as (
    select
        employee_name,
        `Statements`
    from Incorrect_records
    where `Statements` like '%cash%'
)
select
    cs.employee_name,
    cs.`Statements`
from Cash_statements cs
left join Suspect_employees se on cs.employee_name = se.employee_name
where se.employee_name is null;
```
